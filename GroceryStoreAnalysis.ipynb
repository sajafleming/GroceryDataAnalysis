## Bringing the bacon home: from where?
An analysis of various types of grocery store shopping trends from 1983-2013. Data is from table 14 of Food Expenditures on the USDA website:

http://www.ers.usda.gov/data-products/food-expenditures.aspx

```python
import pandas as pd
from statsmodels.api import OLS
pd.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier
%matplotlib inline

import math

import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
mpl.rcParams["figure.figsize"] = (15, 8)
```

```python
spending_df = pd.read_csv('table14.csv', thousands=',').set_index("Year")
spending_df.head()
```

```python
spending_percentages_df = pd.read_csv('percentage.csv').set_index("Year")
spending_percentages_df.head()
```

```python
spending_percentages_df.tail()
```

```python
spending_df[['Supermarkets','Specialty food stores']].plot()
```

### Graph 1: Analysis
We can see from this graph that supermarket sales have an upward trend over time. However, this graph does not do a good job of explaining how specialty food stores sales changed during this time period.

```python
spending_df[['Supermarkets','Specialty food stores']].plot(secondary_y=['Supermarkets'])
```

### Graph 2: Analysis
Using two axes gives us more information about specialty food store trends. It also allows us to compare both supermarkets and specialty stores side by side. Unlike Graph 1, we are able to see that specialty food store sales have increased as well...just not by as much. So how much can we learn from this graph? Well, we would expect to see an upward trend in sales due to inflation and population growth, so a better way to judge the performance of the different store types would be to look at the percentage of total sales and how they vary over time.

```python
spending_percentages_df[['Supermarkets','Specialty food stores','Warehouse clubs and supercenters']].plot(
    secondary_y=['Supermarkets', 'Warehouse clubs and supercenters'])
```

### Graph 3: Analysis
Here I plotted what percent of the total was spent at supermarkets and specialty food stores. I added warehouse clubs and supercenters to help try to explain why the percent of money spent at both supermarkets and speciality food stores was decreasing. In 1983 warehouse clubs and supercenters had 1% of the market. In 2013 they were 16.3% of the market. Sam's club and Costco both opened in 1983 and they have steadily been taking more of the market share. This competition might help exlain why specialty food stores haven't been doing too well. And of course, warehouse clubs and supercenters are competing with supermarkets so the supermarkets are declining (ever so slightly).
